```python
import pandas as pd
from datetime import datetime
```

```python
df_All=pd.read_csv('train.csv')
#df=pd.read_csv('Porto_taxi_data_test_partial_trajectories.csv')
```

```python
pd.options.mode.chained_assignment = None
df=df_All
```

```python
df['TIMESTAMP'] = df['TIMESTAMP'].astype('int')
df['TIMESTAMP']=df['TIMESTAMP'].apply(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
df.head()
```

```python
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
df['TIMESTAMP']=df['TIMESTAMP'].dt.round('15min') 
```

```python
df.head()
```

```python
TEST=df.groupby(['TIMESTAMP']).size().to_frame(name='TAXI_COUNT').reset_index()
```

```python
TEST.sort_values(by='TIMESTAMP')#,ascending=False)
```

```python
OUTPUT=TEST.sort_values(by='TIMESTAMP')
```

```python
OUTPUT.to_csv('out.csv')
```
